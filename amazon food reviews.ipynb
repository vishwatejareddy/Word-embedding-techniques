```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("C:/Users/vishwateja/Desktop/amazon/Reviews.csv")
data.shape
```

```python
#assign values less than 3 to neg and greater than 3 to pos and remove values with 3
final_data=data[data.Score!=3]
def partition(i):
    if i<3:
        return 'Neg'
    else:
        return 'pos'
posneg=final_data['Score'].map(partition)
final_data['Score']=posneg
```

```python
final_data['Score']
final_data.shape
```

```python
#sort data by product id 
sorted_data=final_data.sort_values('ProductId',axis=0,ascending=True)
sorted_data.head()
```

```python
#remove duplicate values
final_data=sorted_data.drop_duplicates(subset={"Id","ProfileName","Time","Text"},keep="first",inplace=False)
final_data=final_data[final_data.HelpfulnessNumerator <= final_data.HelpfulnessDenominator]
final_data.shape
```

```python
#data preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
sno=nltk.stem.SnowballStemmer('english')
def cleanhtml(sentence):
    cleanr=re.compile('<.*?>')
    cleantext=re.sub(cleanr,'',sentence)
    return cleantext
def cleanpun(sentence):
    cleanr=re.sub(r'[?|,|!|\'|"|#]',r'',sentence)
    cleantext=re.sub(r'[)|(|.|,|/|\|:|\"]',r'',cleanr)
    return cleantext
```

```python
import nltk
nltk.download('stopwords')
stop=set(stopwords.words('english'))
stop
```

```python
#datapreprocessing removing html,punctuation,stop words, stemming,tokenizing etc...
pos_words=[]
neg_words=[]
i=0
string=''
finalstring=[]
for sent in final_data['Text'].values:
    filteredsentence=[]
    sent=cleanhtml(sent)
    sent=cleanpun(sent)
    for word in sent.split():
        if (word.isalpha()) & (len(word)>2):
            if (word.lower() not in stop):
                s=sno.stem(word.lower())
                filteredsentence.append(s)
                if final_data['Score'].values[i]=='pos':
                    pos_words.append(s)
                if final_data['Score'].values[i]=='Neg':
                    neg_words.append(s)
            else:
                continue
        else:
            continue
    string=' '.join(filteredsentence)
    finalstring.append(string)
i=i+1
```

```python
final_data['cleanedtext']=finalstring
```

```python
#bag of words
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
countvector=CountVectorizer()
final_counts=countvector.fit_transform(final_data['Text'].values)
```

```python
#tfidf and bigram 
tfidfvector=TfidfVectorizer(ngram_range=(1,2))
final_counts=tfidfvector.fit_transform(final_data['Text'].values)
final_counts
```

```python
#convert sparse matrixto numpy array
import numpy as np
print(final_counts[3,:].toarray()[0])
np.argsort(final_counts[3,:].toarray()[0])[::-1][:25]
```

```python
#top 25 tfidf words
def topidf(row,features,n=25):
    ids=np.argsort(row)[::-1][:n]
    topfeatures=[(features[i],row[i]) for i in ids]
    df=pd.DataFrame(topfeatures)
    df.columns=['feature','tfidf']
    return df
topidf(final_counts[3,:].toarray()[0],features,25)
```

```python
#word2vec
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import pickle
```

```python
import gensim
i=0
listsent=[]
for sent in final_data['Text'].values:
    filtered=[]
    final=cleanhtml(sent)
    final=cleanpun(final)
    for word in final.split():
        if (word.isalpha()):
            filtered.append(word.lower())
        else:
            continue
    listsent.append(filtered)
            
```

```python
w2vmodel=gensim.models.Word2Vec(listsent,min_count=5,size=50,workers=4)
```

```python
words=list(w2vmodel.wv.vocab)
print(len(words))
```

```python
w2vmodel.wv.most_similar('tasty')
```

```python
#avg-w2v and tfidf-w2v
avgw2v=[]
for sent in listsent:
    avgw=np.zeros(50)
    count=0
    for word in sent:
        try:
            vec=w2vmodel.wv[word]
            avgw += vec
            count += 1
        except:
            pass
    avgw /= count
    avgw2v.append(avgw)
```
